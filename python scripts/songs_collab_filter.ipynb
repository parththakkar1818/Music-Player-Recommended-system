```python
import pandas as pd
# from sklearn.preprocessing import LabelEncoder
```

```python
songs = pd.read_csv("../Dataset/songs.csv")
print(songs.head(2))
# songs_df = pd.DataFrame(songs)
```

```python
user_ratings = songs.pivot_table(index=["userId"], columns= "name", values = "rating")
user_ratings
```

```python
item_similarity_df = user_ratings.corr(method="pearson")
item_similarity_df.head()
```

```python
def get_similar_movie(song_name, user_rating):
    similar_score= item_similarity_df[song_name]*(user_rating-5)
    similar_score = similar_score.sort_values(ascending=False)

    return similar_score
```

```python
user = {(songs["name"][i], songs["rating"][i]) for i in songs.index if songs["userId"][i] == 1}
user
```
