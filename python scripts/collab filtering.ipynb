```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings= pd.read_csv("Dataset/toy_dataset.csv", index_col=0)
ratings = ratings.fillna(0)
ratings
```

```python
def standardize(row):
    new_row = (row-row.mean()) / (row.max() - row.min())
    return new_row

ratings_std = ratings.apply(standardize)
ratings_std
```

```python
#we are taking transpose as we want similarity between items which need to be in rows
item_similarity = cosine_similarity(ratings_std.T)
item_similarity
item_similarity_df=pd.DataFrame(item_similarity, index=ratings.columns, columns= ratings.columns)
item_similarity_df
```

```python
def get_similar_movie(movie_name, user_rating):
    similar_score= item_similarity_df[movie_name]*(user_rating-2.5)
    similar_score = similar_score.sort_values(ascending=False)

    return similar_score

print(get_similar_movie("action1",1))
```

```python
action_lover={("action1",5),("romantic2",1),("romantic1",1)}

similar_movies = []

for movie,rating in action_lover:
    similar_movies.append(get_similar_movie(movie,rating))

# print(similar_movies)
similar_movies_df = pd.DataFrame(similar_movies)
print(similar_movies_df)
print(similar_movies_df.sum().sort_values(ascending=False))
```
