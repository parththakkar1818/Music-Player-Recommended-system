```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('hindi_songs.csv')

# Descriptive statistics for the Rating column
print("Descriptive Statistics for Rating:")
print(data['Rating'].describe())

# Visualize the distribution of Rating values
plt.figure(figsize=(8, 6))
plt.hist(data['Rating'], bins=20, edgecolor='black')
plt.title('Distribution of Rating')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```

```python

# Plot pie chart for Genre categories
genre_counts = data['Genre'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Genre Categories')
plt.axis('equal')
plt.show()

```
